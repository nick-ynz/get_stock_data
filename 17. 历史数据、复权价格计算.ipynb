```python
import pandas as pd

pd.set_option('expand_frame_repr', False) # 列不换行
pd.set_option('display.max_rows', 5000) # 最多显示行数

path='sh600000.csv'
df = pd.read_csv(path, encoding='gbk', skiprows=1)
print(df)
```

```python
df['涨跌幅'] = df['收盘价']/df['前收盘价'] - 1 # 真实如果持有股票一天的涨跌幅

df['复权因子'] = (1 + df['涨跌幅']).cumprod() # 第一天买了1块钱股票，到今天会有多少钱

# 后复权价：第一天买入同第一天收盘价一样的股票，之后每天的价格
df['收盘价_后复权'] = df['复权因子']*(df.iloc[0]['收盘价'])/df.iloc[0]['复权因子']

# 前复权价：买的股票价格与最后的收盘价相同，之前每天的价格
df['收盘价_前复权'] = df['复权因子']*(df.iloc[-1]['收盘价'])/df.iloc[-1]['复权因子']

df['最高价_前复权'] = df['最高价'] / df['收盘价'] * df['收盘价_前复权']
df['最低价_前复权'] = df['最低价'] / df['收盘价'] * df['收盘价_前复权']
df['开盘价_前复权'] = df['开盘价'] / df['收盘价'] * df['收盘价_前复权']

print(df)
```
